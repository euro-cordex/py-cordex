# ESGF access

We want to have a slightly more easy access to the ESGF meta data. For this, we tried to abstract away some of the tedious work to access ESGF with python.

```python
import pandas as pd

from cordex import esgf_access

# Euro-Cordex search attributes
attrs = {
    "project": "CORDEX",
    "domain": "EUR-11",
    "variable": "orog",
    "rcm_name": "REMO2015",
}

# create a datasets search result object
results = esgf_access.datasets(attrs, verbose=True)

# creates a pandas dataframe from dataset ids
df = esgf_access.to_pandas(results, columns="CORDEX")
```
